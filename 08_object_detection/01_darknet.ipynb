```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
%%bash
rm -rf /content/*
```

Clone darknet repository

```python
%%bash
git clone https://github.com/pjreddie/darknet.git
cd darknet/
make
```

Get YOLOv3 model weights

```python
%%bash
cd darknet
wget -q https://pjreddie.com/media/files/yolov3.weights
mkdir images
```

Predictions

```python
%%bash
cd darknet
./darknet detect cfg/yolov3.cfg yolov3.weights images/workplace.jpg
```

```python
img = cv2.imread('darknet/predictions.jpg')
cv2_imshow(img)
```
